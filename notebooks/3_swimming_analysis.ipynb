Let's make it again, but grouping all strokes.

```python
def find_strokes(row):
    if 'Freestyle' in row['disciplineName']:
        return 'Freestyle'
    elif 'Breaststroke' in row['disciplineName']:
        return 'Breastroke'
    elif 'Butterfly' in row['disciplineName']:
        return 'Butterfly'
    elif 'Backstroke' in row['disciplineName']:
        return 'Backstroke'
    elif 'Medley' in row['disciplineName']:
        return 'Medley'
    else:
        return 'Unknown'

df_av['Style'] = df_av.apply(find_strokes, axis=1)
```

```python
df_av.head()
```

```python
pivot_strokes = pd.pivot_table(df_av, values='percentage_change', index='year', columns='Style')
pivot_strokes.plot(kind='line', figsize=(10,7))
```

Much cleaner, let's save it

```python
pivot_strokes.to_csv('3_pivot_strokes.csv')
```
