```python
import pandas as pd
from datetime import timedelta
```

```python
df = pd.read_csv('1_output.csv')
df.columns
```

```python
# I will start by selecting the columns I'm most interested in.

df = df[['rank','time','records','resultDate','disciplineName','fullName','participantCountryCode']].copy()
df.head()
```

```python
# Add year column, remove resultDate

df['year'] = df['resultDate'].astype(str).str[:4]
df['year'] = df['year'].astype(int)
df = df.drop('resultDate', axis=1)
df.head()
```

```python
df.dtypes
```

```python
df.time.str.len().value_counts()
```

It seems that the `time` column is formatted in various different ways, and pandas doesn't recognize what it is. To make some calculations, we need to format it to a timedelta object. But to do so, every time must be formatted in the same 11-digit format HH:MM:SS.DD

```python
# Format all times so that they will all have the same 11-digit format HH:MM:SS.DD

df['time'] = df['time'].apply(lambda x: '00:0' + x if len(x) == 7 else x)
df['time'] = df['time'].apply(lambda x: '00:00:' + x if len(x) == 5 else x)
df['time'] = df['time'].apply(lambda x: '00:' + x if len(x) == 8 else x)
df.loc[df['time'].str.len() == 10]
```

```python
df.time.str.len().value_counts()
```

There are still four times that are formatted differently. Because they are just four, we are probably good if we fix them manually.

```python
df.loc[df['time'].str.len() == 10]
```

```python
# Manually correct the four times formatted differently

df.at[5040,'time']='00:01:06.00'
df.at[5062,'time'] = '00:02:23.26'
df.at[17311,'time'] = '00:01:04.07'
df.at[58871,'time'] = '00:05:07.20'
df.time.str.len().value_counts()
```

```python
df['time'] = pd.to_timedelta(df['time'])
df.dtypes
```

```python
df.head()
```

Now I have my timedelta object. However, to do some calculations it's probably easier to have a float. Let's add a column that tells me how many seconds is every time.

```python
df['time_seconds'] = df['time'].dt.total_seconds()
df.tail()
```

Now I probably want to save my clean df to a new csv file, and navigate through it on Excel to see if something seems in order.

```python
df.to_csv('2_output_clean.csv')
```

### After reviewing the data on Excel, I've seen there's a big problem with 800 m freestyle in 2002 !

```python
df.loc[(df['disciplineName'] == 'Men 800 Freestyle') & (df['year'] == 2002)]
```

Positions from 34 to 44 have values that are definetly too high!
It must have been a mistake, those are 1,500m times, not 800. Let's correct it.

```python
df.loc[(df['disciplineName'] == 'Men 800 Freestyle') & (df['year'] == 2002) & (df['time_seconds'] > 900)]
```

```python
# Let's fix this

df.loc[(df['disciplineName'] \
        == 'Men 800 Freestyle') & \
        (df['year'] == 2002) & \
        (df['time_seconds'] > 900), \
        'disciplineName'] = 'Men 1500 Freestyle'
```

```python
len(df.loc[(df['disciplineName'] == 'Men 1500 Freestyle') & (df['year'] == 2002)])
```

```python
df.loc[(df['disciplineName'] == 'Men 800 Freestyle') & (df['year'] == 2002)]
```

### It seems it worked! Let's save our changes

```python
df.to_csv('2_output_clean.csv')
```

Before moving forward with our analysis, we want a standard dataset.
We tried to get the info for the top-50 times of the year, but this info doesn't seem available for all years!
How can we fix this?

```python
df['year'].value_counts()
```

```python
disciplines_year = df.groupby(['disciplineName', 'year']).size().reset_index(name='count')
disciplines_year.sort_values(by='count').head(20)
```

```python
disciplines_year.sort_values('count', ascending=True).head(30)
```

The first thing that I notice is that some disciplines have very little data for less recent years.
That's because they have been introduced pretty recently. Maybe we want to get rid of them?

* 50m butterfly M/W
* 50m breastsroke M/W
* 50m backstrock M/W
* 800m freestyle M
* 1500m freestyle W

```python
disciplines_year.groupby('disciplineName')['count'].sum().sort_values(ascending=False).tail(20)
```

50m is a pretty important race, so I decide to remove from the first race after the Women 50 Freestyle. That means the last eight races.

```python
# Identify disciplines less present in the df and store them in a list

to_remove = disciplines_year.groupby('disciplineName')['count'].sum().sort_values(ascending=False).tail(8)

to_remove = to_remove.index.tolist()
to_remove
```

```python
# Delete rows with incomplete disciplines

df = df[~df['disciplineName'].isin(to_remove)]
df
```

Now that I have restricted my df to only the races that I want to keep, let's see what years should I select to have a consistent dataset.

```python
df['year'].value_counts().sort_index()
```

It seems that data is really missing for years 1989 and 1990! However, I should be probably fine if I just select starting from 2003.

```python
df = df[df['year'] > 2002]
df['year'].value_counts().sort_index()
```

```python
df.groupby('disciplineName').size()
```

Yep. That seems the way to go. I'll do my analysis of the top-50 times for the last 20 years.
